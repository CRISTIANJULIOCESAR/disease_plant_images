```python
from rembg import remove
from PIL import Image
import io
import os
from tqdm import tqdm

# Función para eliminar el fondo de una imagen
def remove_background(input_path, output_path):
    with open(input_path, 'rb') as i:
        input_data = i.read()
        output_data = remove(input_data)
        image = Image.open(io.BytesIO(output_data))
        image.convert("RGB").save(output_path, "JPEG")

# Función para procesar todas las imágenes en una carpeta
def process_folder(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    for file in tqdm(files, desc=f"Processing {input_folder}", unit="image"):
        input_path = os.path.join(input_folder, file)
        output_path = os.path.join(output_folder, os.path.splitext(file)[0] + '.jpg')
        if not os.path.exists(output_folder):
            os.makedirs(output_folder)
        remove_background(input_path, output_path)

# Carpetas para procesar
folders_to_process = [
    'Test/Healthy', 'Test/Leaf Blight', 'Test/Mosaic',
    'Train/Healthy', 'Train/Leaf Blight', 'Train/Mosaic'
]

# Directorio base donde se encuentra la carpeta 'descomprimido'
base_input_dir = '/content/descomprimido/Hojas'
# Directorio base para las carpetas de salida
base_output_dir = '/content/plantas_recortadas'

# Procesa cada una de las subcarpetas
for folder in folders_to_process:
    input_folder = os.path.join(base_input_dir, folder)
    output_folder = os.path.join(base_output_dir, folder)
    process_folder(input_folder, output_folder)


# Comprimir el directorio de salida y moverlo a Google Drive
!zip -r /content/drive/My\ Drive/plantas_recortadas.zip /content/plantas_recortadas
```
